```python
import matplotlib.pyplot as plt
from sklearn import  linear_model 
import seaborn as sns
import pandas as pd 
import numpy as np
import os
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
df = pd.read_csv('framingham.csv', sep =",")
df
#print (df)
#df.plot.scatter(x="sysBP",y="prevalentHyp")
#df.plot.scatter(x="diaBP",y="prevalentHyp")
X_train = np.array(df['sysBP']).reshape((-1, 1))
Y_train = np.array(df['prevalentHyp'])

model = linear_model.LinearRegression()
model.fit(X_train, Y_train)

# parametros
print(f"intercepto (b): {model.intercept_}")
print(f"pendiente (w): {model.coef_}")

#df.plot.scatter(x="glucose",y="TenYearCHD")

w = 0.01463067
b = -1.6258805261061442
x = np.linspace(45,df['sysBP'].max(),100)
y = 1/(1+np.exp(-(w*x+b)))
#prueba del modelo 
personas= [65,120,165,245]
y2 =np.frompyfunc( lambda w, b, x: 1/(1+np.exp(-(w*x+b))),3,1)
probalidades = y2(w,b,personas)
#prueba = np.linspace(45,personas[].max(),100) 

# grafica de la recta
df.plot.scatter(x="sysBP",y="prevalentHyp")
plt.plot(x, y, '-r')
plt.scatter(personas, probalidades, marker="*", c="darkorange", s=50)
plt.text(personas[0],probalidades[0], "%0.2f" % probalidades[0],size=9, color="black")
plt.text(personas[1],probalidades[1], "%0.2f" % probalidades[1],size=9, color="black")
plt.text(personas[2],probalidades[2], "%0.2f" % probalidades[2],size=9, color="black")
plt.text(personas[3],probalidades[3], "%0.2f" % probalidades[3],size=9, color="black")
plt.ylim(-0.1,df['prevalentHyp'].max()*1.1)
plt.grid()
plt.show()



```
