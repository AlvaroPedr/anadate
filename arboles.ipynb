```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score

# se modifico la extrencion de los datos a .csv
df = pd.read_csv('wine.csv')

plt.hist(df.Instances)
plt.show()

predictors_col =["Alcohol", "malicacid","Ash","Alcalinity_ash","magnesium","Total_phenols","Flavanoids","Nonflavanoid_phenols","Proanthocyanins","Color_intensity","Hue","OD280/OD315_diluted_wines","Proline" ]
target_col = ["Instances"]

perdictors = df[predictors_col]
targe = df[target_col]

x_train, x_tes, y_train, y_tes = train_test_split(perdictors, targe, test_size= 0.2, random_state= 40)

tree = DecisionTreeClassifier()
arbol = tree.fit(x_train,y_train)


plt.figure(figsize=(20,12))  # set plot size (denoted in inches)
plot_tree(arbol, fontsize=9)
plt.show()

#plot_tree(arbol)


prueba = arbol.predict(x_tes)
print (pd.crosstab(np.array([y[0]for y in y_tes.values.tolist()]),prueba,rownames=['Acrual'],colnames=['prueba']),'\n\n\n')


accuracy = accuracy_score(y_tes,prueba)
print( "el porcentaje de exactitud es %% %s"%(accuracy))
plt.show()
```
